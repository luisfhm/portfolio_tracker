```python
import requests
from datetime import datetime

# Configuración
token = "86cf2a6f5b2398710af3c1fecce92e"          # tu token
inicio = "2026-01-02"
final  = "2026-01-02"                             # noviembre 2025 (30 días)
base_url = "https://api.databursatil.com/v2/intradia"

# Lista de emisoras a probar (agrega/quita las que quieras)
emisoras = [
    "CEMEXCPO",
    "FEMSAUBD",
    "WALMEX*",
    "NVDA*",
    "GMEXICOB",
    "GFNORTEO",
    "GAPB",
    # "AC",           # Arca Continental
    # "BIMBOA",       # Bimbo
    # "KOFUBL",       # Coca-Cola FEMSA (a veces varía)
]

print(f"Consultando datos históricos del {inicio} al {final}\n")

for emisora in emisoras:
    params = {
        "token": token,
        "intervalo": "1m",
        "inicio": inicio,
        "final": final,
        "emisora_serie": emisora,
        "bolsa": "BMV"
    }
    
    try:
        response = requests.get(base_url, params=params, timeout=10)
        
        if response.status_code == 200:
            data = response.json()
            num_dias = len(data)
            
            if num_dias > 0:
                # Suponiendo formato { "fecha": [precio_cierre, volumen], ... }
                fechas = sorted(data[emisora].keys())
                primer_dia = fechas[0]
                ultimo_dia = fechas[-1]
                precio_inicio = data[emisora][primer_dia]
                precio_final  = data[emisora][ultimo_dia]
                
                print(f"✓ {emisora:10} → OK | {num_dias:2} días | {primer_dia} ${precio_inicio:.2f} → {ultimo_dia} ${precio_final:.2f}")
                # Opcional: mostrar todos los datos
                # for fecha, valores in sorted(data.items()):
                #     print(f"  {fecha}: ${valores[0]:.2f} | vol {valores[1]:,.0f}")
            else:
                print(f"✗ {emisora:10} → OK pero sin datos")
        else:
            print(f"✗ {emisora:10} → Error {response.status_code} - {response.text[:150]}...")
            
    except requests.exceptions.RequestException as e:
        print(f"✗ {emisora:10} → Falló la conexión: {e}")
    
    print("-" * 60)
```
